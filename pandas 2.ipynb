```python
import pandas as pd
df = pd.read_csv('datasets/sample2.csv', index_col =['Roll No.'])
df.head(10)
```

##### loc is label-based, which means that you have to specify rows and columns based on their row and column labels.
##### iloc is integer position-based, so you have to specify rows and columns by their integer position values (0-based integer position).

```python
df.loc[3]
```

```python
df.loc[[5,10,20,25]]
```

```python
df.loc[6, 'Branch']
```

```python
df.loc[df['Physics']<=50, ['Maths']]
```

## iloc

```python
df.iloc[0]
```

```python
df.iloc[[0,1, 25,6]]
```

```python
df.iloc[:,2]
```

```python
branch_group.groups
```

```python
# Merge() Function
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Physics':[34,67,34,89,12]})
df1
```

```python
df2 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Chemistry':[78,33,39,81,90]})
df2
```

```python
pd.merge(df1, df2, on = 'Roll No.') 
```

```python
pd.merge(df1, df2)   # on - by default intersecting column is taken
```

```python
# append() Function
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Maths':[45,78,45,90,66], 'Physics':[33,67,12,90,44]})
df2 = pd.DataFrame({'Roll No.':[6,7,8,9,10], 'Maths':[78,73,45,90,69], 'Physics':[23,67,88,0,98]})
print(df1)
print(df2)
```

```python
df1.append(df2, ignore_index=True)
```
