```python
#import library
import pandas as %pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
#reading csv
df = pd.read_csv('../datasets/zomato.csv')
df.head()
```

```python
df =df.drop(['url', 'address', 'phone', 'dish_liked', 'reviews_list', 'menu_item'], axis=1)
df
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
df.info()
```

```python
df['rate'].unique()
```

# removing 'NEW' and '-'

```python
#made function to remove all unwanted string and '/5' we need only rating not out of
def handlerate(value):
    if(value == 'NEW' or value == '-'):
        np.nan
    else:
        value=str(value).split('/')      
        return float(value[0])
        
        
df['rate']=df['rate'].apply(handlerate)
df['rate'].unique()
```

```python
df.rate.isnull().sum()
```

### filling null values with means values

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
#inplace meaning make changes at the original place. Don't need to create new rows
```

```python
df['rate'].isnull().sum()
```

```python
df.info()
```

# if they are not having many null values thenb just drop that 

```python
df.rename(columns = {'approx_cost(for two people)' : 'cost2plates', 'listed_in(type)' : 'type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

### we have to keep only one columns for location between location and listed_in(city)



```python
df = df.drop(['listed_in(city)'], axis=1)
```

```python
df
```

```python
df['cost2plates'].unique()
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
    

df['cost2plates']=df['cost2plates'].apply(handlecomma)
df['cost2plates'].unique()
```

```python
df.head()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_typeslessthan1000 = rest_types[rest_types<1000]
rest_typeslessthan1000
```

```python
def handle_rest_type(value):
    if value in rest_typeslessthan1000:
        return 'others'
    else:
        return value
       
    
    
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
### cleaning location
```

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = False)
location_lessthan300 = location[location<300]
location_lessthan300
```

```python
def handle_location(value):
    if value in location_lessthan300:
        return 'others'
    else:
        return value
    
    
    
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)
```

```python
cuisines_lessthan300 = cuisines[cuisines<300]
cuisines_lessthan300
```

```python
def handle_cuisines(value):
    if value in cuisines_lessthan300:
        return 'others'
    else:
        return value
    
    
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
